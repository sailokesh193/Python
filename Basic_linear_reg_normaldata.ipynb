```python
#LINEAR REGRESSION
from statistics import mean
```

```python
import numpy as np
import pandas as pd
```

```python
import  matplotlib.pyplot as plt
#for getting grid lines we may use this below one
from matplotlib import style
style.use('fivethirtyeight')
```

```python
xs=np.array([1,2,3,4,5,6,7,8],dtype=np.float64)
ys=np.array([3,4,6,8,7,9,7,8],dtype=np.float64)
```

```python
plt.scatter(xs,ys)
plt.show()
```

```python
#HERE WE ARE DEFINING THE FUNCTION
def best_fit_slope(xs,ys):
    m=( ((mean(xs)*mean(ys))-mean(xs*ys)) / ((mean(xs)**2)- (mean(xs**2))) )
    return m
```

```python
m=best_fit_slope(xs,ys)
```

```python
m
```

```python
#HERE WE ARE CALCULATING THE INTERCEPT
def best_intercept(xs,ys):
    b=mean(ys)-m*mean(xs)
    return b
```

```python
b=best_intercept(xs,ys)
```

```python
b
```

```python
#best fit line
regression_line=[(m*x)+b for x in xs]
#we can also do in this way
#for x in xs
#   regression_line.append((m*x)+b)
```

```python
plt.scatter(xs,ys)
plt.plot(xs,regression_line)
plt.show()
```

```python
#for simple prediction for single value
predict_x=8
predict_y=((m*predict_x)+b)
```

```python
plt.scatter(xs,ys)
plt.plot(xs,regression_line)
plt.scatter(predict_x,predict_y,color="red")
plt.show()
```

```python
#Now it is predicting for group of values.
predict_x=[8,5,3.3,6,5.99,3.98,7,5.65]
predict_y=[((m*x)+b) for x in (predict_x)]
```

```python
plt.scatter(xs,ys)
plt.plot(xs,regression_line)
plt.scatter(predict_x,predict_y,color="red")
plt.show()
```

```python
#testing accuracy
def squared_error(ys_orig,ys_line):
    return sum((ys_line-ys_orig)**2)
```

```python
def coefficient_of_determination(ys_orig,ys_line):
    y_mean_line=[mean(ys_orig) for y in ys_orig]
    squared_error_regr=squared_error(ys_orig,ys_line)
    squared_error_y_mean=squared_error(ys_orig,y_mean_line)
    return 1-(squared_error_regr/squared_error_y_mean)
```

```python
r_squared=coefficient_of_determination(ys,regression_line)
```
