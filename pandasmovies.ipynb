```python
#distplot is like histogram and it also give trends
sns.distplot(stats['internetusers'])
show()
```

```python
sns.distplot(stats['internetusers'],label='internetusers',color='green')
plt.legend()
show()
```

```python
#here we are adding labels and number of bins
sns.distplot(stats['internetusers'],label='internetusers',color='green',bins=30)
plt.legend()
show()
```

```python
#plotting birth rate
sns.distplot(stats['birthrate'],label='birthrate',color='red',bins=50)
plt.legend()
show()
```

```python
vis2=sns.boxplot(data=stats,x='incomegroup',y='birthrate')
vis2
show()
```

```python
#best fit line
vis3=sns.lmplot(data=stats,x='internetusers',y='birthrate')
show()
```

```python
#here iam removing best fit line
vis3=sns.lmplot(data=stats,x='internetusers',y='birthrate',fit_reg=False)
show()
```

```python
vis4=sns.lmplot(data=stats,x='internetusers',y='birthrate',hue='incomegroup',fit_reg=False,size=8)
show()
```

```python
vis4=sns.lmplot(data=stats,x='internetusers',y='birthrate',fit_reg=False,hue='incomegroup',aspect=1)
show()
```

```python
vis4=sns.lmplot(data=stats,x='internetusers',y='birthrate',fit_reg=False,hue='incomegroup',aspect=0.5)
show()
```
